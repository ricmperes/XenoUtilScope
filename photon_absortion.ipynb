# Photon absorption processes plot

Data from XCOM: Photon Cross Sections Database: https://www.nist.gov/pml/xcom-photon-cross-sections-database

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/photon_absorption_xenon.txt',
                   sep = ' ', 
                   names = ['energy','scat_inc','scat_cohe','photoelec',
                            'pair_prod_nuclear','pair_prod_elect','total',
                            'total_no_coherent'],
                   header = None,
                   skiprows=4,
                   index_col=False
                  )
```

```python
for _col in data.columns.values[1:]:
    plt.plot(data['energy']*1e3, data[_col], label = _col)
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.show()
```

```python
plt.plot(data['energy']*1e3, data['total'], label = 'Total')
plt.plot(data['energy']*1e3, data['photoelec'], label = 'Photoelectric effect')
plt.plot(data['energy']*1e3, data['scat_cohe'], label = 'Compton scattering')
plt.plot(data['energy']*1e3, data['pair_prod_nuclear'], label = 'Pair production in nucl. field')
plt.plot(data['energy']*1e3, data['pair_prod_elect'], label = 'Pair production in electron field')

plt.xlabel('Photon energy [keV]')
plt.ylabel('Absorptoin coefficient [cm$^2$/g]')
plt.xlim(1,1e5)
plt.ylim(1e-4,1e5)
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.show()
```
